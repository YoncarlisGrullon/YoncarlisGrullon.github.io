```python
#Import numpy
import numpy as np

#Import pandas
import pandas as pd

#Import package
import matplotlib.pyplot as plt

HealthUs2021 = pd.read_csv('Health by U.S. County 2021 (v.sp22) - County Health.csv')
display(HealthUs2021)

# % Adults with Obesity, AWO
AWO = HealthUs2021['% Adults with Obesity'].to_numpy()

# % Physically Inactive, PI
PI = HealthUs2021['% Physically Inactive'].to_numpy()

#plot graph
plt.title('% Adults with Obesity In Us Countys and Physically Inactive')
plt.xlabel('% Adults with Obesity in Us Counties')
plt.ylabel('% Physically Inactive')
#plot
plt.plot(AWO, PI)
plt.show()

#histagram
plt.hist(AWO,30 , edgecolor = 'black', fill = True, facecolor='Blue', cumulative=False)
plt.ylabel('Number of Us Counties')
plt.xlabel('Percentage Of Adult with Obesity in Us Counties and Physically Inactive', )
plt.hist(PI,30 , edgecolor = 'black', fill = True, facecolor='Red', cumulative=False, alpha = 0.6)
plt.show()

# Sample data
data1 = AWO
data2 = PI

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Histogram 1
ax1.hist(data1, bins=50, color='blue', alpha=0.7)
ax1.set_xlabel('% Adult with Obesity in Us Counties')
ax1.set_ylabel('Number of Us Counties')

# Histogram 2
ax2.hist(data2, bins=50, color='red', alpha=0.7)
ax2.set_xlabel('% Physically Inactive')
ax2.set_ylabel('Number of Us Counties')

plt.tight_layout()
plt.show()





#stair graph
plt.stairs(AWO,edgecolor ='blue', facecolor='Blue')
plt.stairs(PI, edgecolor ='red', facecolor='Red')
plt.xlabel('Number of Us Counties')
plt.ylabel('Percentage Of Adult with Obesity in Us Counties and Physically Inactive')
plt.show()


#generating a scatter plot
plt.scatter(AWO, PI)
plt.show()










```
